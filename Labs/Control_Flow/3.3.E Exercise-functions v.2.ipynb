#### Q3. 
Write a function that takes a collection, converts it to a `tuple` with all duplicates removed and returns the result.

```python
# your code here
def no_dup(collection):
    new = set(collection)
    new = tuple(new)
    return new

no_dup(['a', 'b', 'a', 'a', 'b', 'c'])
```

##### Q5.C
Write a function that accepts one parameter, a list
The function should return the sum total of all numbers in the list, using a loop.
\
*Hint*: Do we want a for loop or a while loop for this kind of question?

```python
# your code here
def sum_list(l):
    s = 0
    for i in range(len(l)):
        s = s + l[i]

    return s
```

```python
my_list=[3,7,1,9]

sum_list(my_list)
```

```python
# A. Write a function that finds the longest string in a given list of strings

# your code here

def longest_string_finder(s):
    i = 0
    longest = s[i]
    for i in range(len(s) -1) :
        if i == len(s):
            longest = s[i]
        elif len(longest) < len(s[i + 1]):
            longest = s[i + 1]
    return longest
```

```python
s=["dfaf","adfaffaf","grrtsrfsgfgfsgsgf","adfaaa"]
b= ["aaaa", "bbb", "cc"]

longest_string_finder(b)
```

```python
# D. Write a function to count how many times a given string contains a given letter

# your code here
def char_count(s2,letter):
    ocrr = 0
    for i in range(len(s2)):
        if s2[i] == letter:
            ocrr = ocrr + 1

    return ocrr
    
```

```python
s2="This is a test about data science short course with excellent students"

char_count(s2,"i")
```
